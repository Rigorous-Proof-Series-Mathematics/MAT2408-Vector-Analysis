## Linearity of the Dot Product

For all $u,v,w \in \mathbb{R}^n$ and $\alpha,\beta \in \mathbb{R}$,

$$
(\alpha u+\beta v)\cdot w=\alpha(u\cdot w)+\beta(v\cdot w).
$$

<!-- proof:start -->

### Proof

By definition,

$$
(\alpha u+\beta v)\cdot w
=
\sum_{i=1}^{n}(\alpha u_i+\beta v_i)w_i.
$$

Distributing gives the result.

<!-- block:end -->

<!-- block:start definition label=def-norm -->

## Norm

The norm of $v \in \mathbb{R}^n$ is

$$
\|v\|=\sqrt{v\cdot v}.
$$

<!-- block:end -->

## Zero Dot Product with Itself

If $v\cdot v=0$, then $v=0$.

<!-- proof:start -->

### Proof

Since

$$
v\cdot v=\sum_{i=1}^{n}v_i^2,
$$

all terms must vanish.
