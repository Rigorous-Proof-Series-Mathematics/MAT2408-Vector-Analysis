## Theorem: Linearity

For all $u,v,w \in \mathbb{R}^n$ and $\alpha,\beta \in \mathbb{R}$,

$$
(\alpha u+\beta v)\cdot w=\alpha(u\cdot w)+\beta(v\cdot w).
$$

<!-- proof:start -->

### Proof

By the definition of the dot product,

$$
(\alpha u+\beta v)\cdot w
=
\sum_{i=1}^n(\alpha u_i+\beta v_i)w_i.
$$

Distributing the terms proves the result.

<!-- block:end -->

<!-- block:start definition -->

## Definition: Norm

The norm of a vector $v \in \mathbb{R}^n$ is defined by

$$
\|v\|=\sqrt{v\cdot v}.
$$

<!-- block:end -->

<!-- block:start lemma -->

## Lemma: Zero Vector

If $v\cdot v=0$, then $v=0$.

<!-- proof:start -->

### Proof

Since

$$
v\cdot v=\sum_{i=1}^n v_i^2,
$$

each term must be zero. Hence $v_i=0$ for all $i$.

<!-- block:end -->

<!-- block:start corollary -->

## Corollary: Positivity of the Norm

For every $v\in\mathbb{R}^n$,

$$
\|v\|\geq 0.
$$

<!-- proof:start -->

### Proof

Since $\|v\|=\sqrt{v\cdot v}$ and $v\cdot v\geq 0$, the result follows.

<!-- block:end -->

<!-- block:start axiom -->

## Axiom: Additive Identity

For every vector $v$ in a vector space $V$, there exists a vector $0\in V$ such that

$$
v+0=v.
$$

<!-- block:end -->

<!-- block:start proposition -->

## Proposition: Symmetry of the Dot Product

For all $u,v\in\mathbb{R}^n$,

$$
u\cdot v=v\cdot u.
$$

<!-- proof:start -->

### Proof

By definition,

$$
u\cdot v=\sum_{i=1}^n u_i v_i.
$$

Since multiplication in $\mathbb{R}$ is commutative, $u_i v_i=v_i u_i$.

<!-- block:end -->

<!-- block:start example -->

## Example: Dot Product in $\mathbb{R}^3$

Let $u=(1,2,3)$ and $v=(4,5,6)$. Then

$$
u\cdot v=1\cdot4+2\cdot5+3\cdot6=32.
$$

<!-- block:end -->

<!-- block:start remark -->

## Remark: Geometric Meaning

The dot product also satisfies

$$
u\cdot v=\|u\|\|v\|\cos\theta.
$$

So it measures how much one vector points in the direction of another.
