## Theorem: Linearity of the Dot Product

Let $u, v, w \in \mathbb{R}^n$ and let $\alpha, \beta \in \mathbb{R}$. Then

$$
(\alpha u + \beta v)\cdot w
=
\alpha(u\cdot w)+\beta(v\cdot w).
$$

### Proof

We write the vectors in coordinates:

$$
u=(u_1,\dots,u_n), \qquad
v=(v_1,\dots,v_n), \qquad
w=(w_1,\dots,w_n).
$$

Then

$$
\alpha u+\beta v
=
(\alpha u_1+\beta v_1,\dots,\alpha u_n+\beta v_n).
$$

Therefore,

$$
(\alpha u+\beta v)\cdot w
=
\sum_{i=1}^{n}(\alpha u_i+\beta v_i)w_i.
$$

Using distributivity in $\mathbb{R}$,

$$
\sum_{i=1}^{n}(\alpha u_i+\beta v_i)w_i
=
\sum_{i=1}^{n}\alpha u_iw_i
+
\sum_{i=1}^{n}\beta v_iw_i.
$$

Since $\alpha$ and $\beta$ are constants,

$$
\sum_{i=1}^{n}\alpha u_iw_i
+
\sum_{i=1}^{n}\beta v_iw_i
=
\alpha\sum_{i=1}^{n}u_iw_i
+
\beta\sum_{i=1}^{n}v_iw_i.
$$

Thus,

$$
(\alpha u+\beta v)\cdot w
=
\alpha(u\cdot w)+\beta(v\cdot w).
$$

This proves the theorem.
